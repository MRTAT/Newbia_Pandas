
# LAB 1


```python
np.random.seed(42)

# Create a random Series of employee salaries (for example, between $50,000 and $120,000)
salaries = pd.Series(np.random.randint(50000, 120000, 20))

print("Employee Salaries:")
print(salaries)
```

```python
# Calculate Mean
mean_salary = salaries.mean()

# Calculate Median
# Giá trị nằm ở giữa của một tập dữ liệu khi các giá trị đã được sắp xếp theo thứ tự tăng dần hoặc giảm dần.
# Median giúp xác định vị trí trung tâm của dữ liệu
median_salary = salaries.median()


# Calculate Standard Deviation
std_dev_salary = salaries.std()

# Calculate Minimum and Maximum
min_salary = salaries.min()
max_salary = salaries.max()



# Calculate Coefficient of Variation  = STD / MEAN

coefficient_of_variation = std_dev_salary / mean_salary

# Display the results
print("\nStatistics:")
print(f"Mean Salary: ${mean_salary:.2f}")
print(f"Median Salary: ${median_salary:.2f}")
print(f"Standard Deviation: ${std_dev_salary:.2f}")
print(f"Coefficient of Variation: {coefficient_of_variation:.2f}%")
print(f"Minimum Salary: ${min_salary:.2f}")
print(f"Maximum Salary: ${max_salary:.2f}")
```

# LAB 2

```python
# Creating a Series for sales data

df = pd.Series([500, 600, 400, 800, 700], index=['Product A', 'Product B', 'Product C', 'Product D', 'Product E'])

df
```

```python
# 1. Descriptive statistics
# We can quickly get an overview of the sales data by applying basic statistical methods.
# Total sales
total_sales = df.sum()
print("Total Sales:", total_sales)

# Average sales
average_sales = df.mean()
print("Average Sales:", average_sales)

# Maximum sales
max_sales = df.max()
print("Maximum Sales:", max_sales)

# Minimum sales
min_sales = df.min()
print("Minimum Sales:", min_sales)
```

```python
# 2. Filtering data
# We can filter the data to find out which products have sales greater than 600.
high_sales = df[df > 600]
print("Products with sales > 600:")
print(high_sales)
```

```python
# 3. Data manipulation
# Let's assume there's a sale going on, and we want to increase the sales of all products by 10%.
sales_data_after_discount = df * 1.1
print("Sales data after a 10% increase:")
print(sales_data_after_discount)
```

```python
# 4. Ranking products based on sales
# To rank the products by sales, you can use the rank() function.
sales_rank = df.rank(ascending = False)
print("Ranking of products by sales:")
print(sales_rank)
```

```python
# 5. Visualizing with y = Sales & x = Index

import matplotlib.pyplot as plt


```

```python
df.plot(kind='bar', title = 'Lab 2') # create name
plt.ylabel('Sales')  # create name
plt.show()
```

```python
# Creating a DataFrame with employee data
data = {
    'Employee ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Cuong', 'Ly', 'Long', 'Vanh', 'Minh', 'Son', 'Khanh', 'Phuc', 'Truong', 'An'],
    'Department': ['HR', 'Finance', 'IT', 'IT', 'Marketing', 'Finance', 'HR', 'Marketing', 'IT', 'HR'],
    'Salary': [70000, 80000, 65000, 70000, 72000, 75000, 68000, 71000, 62000, 65000],
    'Age': [24, 30, 22, 29, 35, 40, 32, 31, 28, 26],
    'Years of Experience': [2, 5, 1, 3, 7, 10, 4, 6, 2, 3],
    'Performance Score': [8, 7, 9, 6, 7, 8, 5, 9, 6, 7],
    'Location': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Chicago', 'New York', 'Los Angeles', 'New York', 'Houston', 'Chicago']
}

df = pd.DataFrame(data)
df
```

```python
# 1. Descriptive Statistics
# Get basic descriptive statistics for the numerical columns to understand the dataset.
# Get descriptive statistics

statistics_summary = df.describe()
print("\nDescriptive Statistics:")
statistics_summary
```

```python
# 2. Grouping by Department and Cal Average Salary
# Calculate the average salary and performance score for each department.

# Grouping by Department and calculating average salary and performance score
dept_statistics = df.groupby('Department')[['Salary', 'Performance Score']].mean()

print("\nDepartment Statistics (Average Salary and Performance Score):")
print(dept_statistics)
```

```python
# 3. Filtering Employees by Performance
# Find employees with a performance score above 7 and salary above 70000.

high_performers = df[(df['Performance Score'] > 7) & (df['Salary'] > 70000)]
print("\nHigh Performers (Score > 7 and Salary > 70000):")
print(high_performers[['Name', 'Department', 'Salary', 'Performance Score']])
```

# Correlation with many values

```python
# 4. Correlation Analysis
# Check the correlation between years of experience, salary, and performance score.
# Calculate correlation matrix

# Xét ngoài đường chéo chính (vì chéo chính là tương quan chính nó)
# NX: Tương quan giữa lương và Số năm kinh nghiệm -> cao (0.65) , còn lại thấp

correlation_matrix = df[['Salary', 'Years of Experience', 'Performance Score']].corr()
print("\nCorrelation Matrix:")
correlation_matrix
```

```python
# 5. Count of Employees in Each Location
# Count the number of employees in each location.

location_count = df['Location'].value_counts().reset_index()
location_count.columns = ['Location', 'Employee Count']
print("\nEmployee Count by Location:")
location_count
```

```python
# 7. Advanced Query: Performance Improvement Analysis
# Identify employees who have a salary below the average salary of their respective department
# but have a performance score above the average performance score for that department.
# Mục đích tăng lương cho những người đã cố gắng

avg_dept_stats = df.groupby('Department')[['Salary', 'Performance Score']].mean().reset_index()


# Đây là một danh sách mới chứa tên các cột mà bạn muốn gán cho DataFrame -> new table
avg_dept_stats.columns = ['Department', 'Avg Salary', 'Avg Performance Score']


# Merge with original dataframe (join: new table <-> DataFrame On Department)
# Table Contains ['Department', 'Avg Salary', 'Avg Performance Score'  + ... Table of Dataframe ]

merged_df = pd.merge(df, avg_dept_stats, on='Department')




# Condition: salary < average salary (ever department) And performance score > the average performance score (ever department)
performance_improvement_candidates = merged_df[(merged_df['Salary'] < merged_df['Avg Salary']) & (merged_df['Performance Score'] > merged_df['Avg Performance Score'])]


print("\nCandidates for Performance Improvement Analysis:")
print(performance_improvement_candidates[['Name', 'Department', 'Salary', 'Performance Score']])
```
